```python
from natasha import (
    Segmenter,
    NewsEmbedding,
    NewsNERTagger,
    Doc
)

segmenter = Segmenter()
emb = NewsEmbedding()
ner_tagger = NewsNERTagger(emb)

def parseText(text):
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)

    result = ''
    per = list(filter(lambda x: x.type == 'PER', doc.spans))
    org = list(filter(lambda x: x.type == 'ORG', doc.spans))
    spans = per + org

    for span in spans:
        for token in span.tokens:
            if hasattr(token, 'start'):
                result += str(token.start)
            else:
                result += str(0)

            result += ' ' + str(len(token.text))
            if span.type == 'PER':
                result += ' ' + 'PERSON '

            if span.type == 'ORG':
                result += ' ' + 'ORG '

    result += 'EOL'
    return result
```

```python
items = []

with open("./Data/dataset_40163_1.txt", "r", encoding="utf-8") as a_file:
    content = a_file.read()
    lines = content.splitlines()
    
    for line in lines:
        items.append(parseText(line))

print(items)
with open('./Data/result.txt', 'w') as f:
    for item in items:
        f.write(item + '\n')
```
