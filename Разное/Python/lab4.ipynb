```python
import numpy as np
import pandas as pd
"""задание 1"""
"""series"""
data = pd.Series(["Вася", "Петя", "Коля"])
print("Объект Series:")
print(data)
"""dataframe"""
car = pd.DataFrame({
    'mark': ['Audi', 'BMW', 'Toyota'],
    'model': ['XL', 'X9s', 'M5'],
    'speed': ['150', '210', '180']
})
print("\nОбъект DataFrame:")
print(car)

"""задание 2"""
data = pd.read_csv('C:/R/titanic.csv')
print("\nСчитали данные из файла titanic.csv:")
print(data)
print("\nГруппировка по PassengerID:")
print(data.groupby(['Sex', 'Survived']) ['PassengerID'].count())
print("\nГруппировка по PClass:")
print(data.groupby(['Sex', 'Survived']) ['PClass'].count())
print("\nСводная таблица:")
pvt = data.pivot_table(index=['Sex'], columns=['PClass'], values='Name', aggfunc='count')
print(pvt)

"""задание 3"""
data = pd.read_csv('C:/R/apple.csv', index_col='Date', parse_dates=True)
print("\nСчитали данные из apple.csv:")
print(data)
data = data.sort_values('Date')
print("\nОтсортировали данные:")
print(data)
print('Средняя цена акции на закрытии =', np.mean(data['Close']))
print('Средняя цена акции с февраля 2012 по февраль 2015 = ' ,data.loc['2012-Feb':'2015-Feb', 'Close'].mean())
print('Средняя цена акции по неделям: \n' , data.resample('W')['Close'].mean())
```
