```python
import numpy as np
import pandas as pd
arr = np.array([1, 2, 3])
"""заданние 1"""
"""инициализация массива"""
array = np.random.randint(1000, 5000, 100)
print("Дан массив:")
print(array)
"""вывод последнего элемента"""
print("Последний элемент данного массива:")
print(array[-1])
"""вывод нечетных элементов"""
print("Нечетные элементы массива:")
answer = (array[array % 2 == 1])
print(answer)
"""вывод четных"""
print("Четные элементы массива:")
answer = (array[array % 2 == 0])
print(answer)
"""сумма элементов массива"""
print("Сумма элементов массива:")
print(np.sum(array))
"""умножаем массив на 5 и прибавляем 3"""
print("Результат умножения массива на 5 и прибавления 3:")
print(array * 5 + 3)

"""задание 2"""
"""загрузка данных из файла"""
data = pd.read_csv('C:\R\cities.csv')
lons = np.array(data['lon'])
print("Массив lons, загруженный из файла cities.csv:")
print(lons)
"""вычисление среднего"""
print("Среднее:")
mean = np.array(lons).mean()
print(mean)
"""вывод значений больше среднего"""
print("Вывод значений из массива больше среднего значения:")
answer = lons[lons >= mean]
print(answer)
"""сортировка"""
print("Сортировка массива:")
print(np.sort(lons))

"""задание 3"""
"""загрузка данных из файла"""
data2 = pd.read_csv("C:/R/train.csv")
ages = np.array(data2['Age'])
print("Массив ages, загруженный из файла train.csv:")
print(ages)
"""вычисление среднего, минимального, максимального, стандартного отклонения, квантилей, медианы"""
print("Среднее:")
print(np.nanmean(ages))
print("Минимальное:")
print(np.nanmin(ages))
print("Максимальное:")
print(np.nanmax(ages))
print("Стандартное отклонение:")
print(np.nanstd(ages))
print("Квантили:")
print(np.nanquantile(ages, 0.5))
print("Медиана:")
print(np.nanmedian(ages))
```
