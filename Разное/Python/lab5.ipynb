```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
''' 1'''
data = pd.read_csv('C:/R/apple.csv')
data = data.sort_values('Date')
print(data)
x = np.array(data['Date'])
y = np.array(data['Low']) 
plt.plot(x, y, label=' Low    ', color='red', linestyle='dashed')
plt.legend()
plt.show()

''' 2'''
x = np.array(data['Date'])
y = np.array(data['Low'])
plt.plot(x, y, 'o', markersize = 4)
plt.xlim(0, 20)
plt.xticks(rotation=45) 
plt.annotate('  Low', xy = (10, 600), xytext=(20, 800), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

x = np.array(data['Date'])
y = np.array(data['High'])
plt.plot(x, y, 'o', markersize = 4)
plt.xlim(0, 20)
plt.xticks(rotation=45) 
plt.annotate('  High', xy = (10, 600), xytext=(20, 800), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()


''' 3'''
def f(x, y):
    return (x + y)

x = np.array(data['Low'])
y = np.array(data['High'])


X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 30, cmap='binary')
```
